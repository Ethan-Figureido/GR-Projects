# Project 2: Geodesic Calculator

This project is to build intuition for how curvature affects motion, and for the mathematics of geodesics and the christoffel coefficients in different metrics.

I want this project to take in a metric (3D and 4D), and an initial position and velocity (our initial conditions), to then output a solution for a geodesic in that metric using a christoffel coefficient.

Ultimately, I would like to visualize the following:
1. orbitting planets
2. falling observers
3. photon trajectories

The relevant equations are:

### Christoffel Connection Coefficient:

$$\Gamma^{\sigma}_{\mu\nu} = \frac{1}{2}g^{\sigma\rho}(\partial_{\mu}g_{\nu\rho} + \partial_{\nu}g_{\rho\mu} - \partial_{\rho}g_{\mu\nu})$$

### Geodesic Equation:

$$\frac{d^{2}x^{\mu}}{d\lambda^{2}} + \Gamma^{\mu}_{\rho\sigma}\frac{dx^{\rho}}{d\lambda} \frac{dx^{\sigma}}{d\lambda} = 0$$

Where $g^{\mu\nu}$ is the given metric, its inverse $g_{\mu\nu}$,  and $x^{\mu}(\lambda)$ is a path parameterized by some parameter $\lambda$. Any path satisfying the geodesic equation is a geodesic.
