The relevant equations are:

### Christoffel Connection Coefficient:

$$\Gamma^{\sigma}_{\mu\nu} = \frac{1}{2}g^{\sigma\rho}(\partial_{\mu}g_{\nu\rho} + \partial_{\nu}g_{\rho\mu} - \partial_{\rho}g_{\mu\nu})$$

### Geodesic Equation:

$$\frac{d^{2}x^{\mu}}{d\lambda^{2}} + \Gamma^{\mu}_{\rho\sigma}\frac{dx^{\rho}}{d\lambda} \frac{dx^{\sigma}}{d\lambda} = 0$$

Where $g^{\mu\nu}$ is the given metric, $g_{\mu\nu}$ its inverse, and $x^{\mu}(\lambda)$ is a path parameterized by some parameter $\lambda$. Any path satisfying the geodesic equation is a geodesic.
